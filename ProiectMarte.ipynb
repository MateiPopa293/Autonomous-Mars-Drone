# **1.1 The Latency Paradox**



---

The core constraint of this mission is the Communication Gap.


*   *The Reality:* The storm is 12 minutes away. A round-trip signal to Earth takes 14 minutes.
*   *The Implication:* If we "call home" for instructions, by the time Earth’s command reaches us, we will have been inside the storm for 2 minutes.

*   *The Decision:* Autonomy is not a feature; it is a survival requirement. The aircraft must act as a Sovereign Agent.






# **1.2 Identifying the "Unknowns" (Critical Questions)**

---

##Before designing a solution, I need to address the gaps in our data. If I were in the mission briefing room, I would ask:



 **A. Sensor Reliability (Signal vs. Noise)**



*   *Consensus:* How many sensors are detecting the storm? Is it a single LiDAR reading or a cross-sensor confirmation (Optical + Barometric)?
*  *Degradation:* Does the dust interfere with our internal navigation or just our external vision?
*   *Ground Truth:* Do we have access to historical weather patterns for this specific Martian quadrant?


**B. Environmental & Technical Constraints**

  

*   *The "Point of No Return":* What is the latest possible second we can execute a maneuver before the aerodynamic stress becomes terminal?

*   *Energy Economics:* How much battery does a 20% course deviation consume? Is a "Safe Landing" more energy-efficient than a "High-Altitude Bypass"?
*   *Cargo Value:* Is the payload "Mission Critical" (medical supplies for a colony) or "Replaceable" (basic sensors)? This dictates the risk appetite.




# **1.3 Strategic Assumptions**

---

##To build a robust model, I am making the following engineering assumptions:


*   *Assumption 1:* The aircraft is equipped with On-board Edge Computing capable of running real-time probabilistic simulations.
*   *Assumption 2:* We have a pre-loaded Topographical Safety Map indicating flat, rock-free zones for emergency landings.
*  *Assumption 3:* Martian dust is potentially Electrostatic, meaning we must assume a total comms blackout once the storm hits.




# **1.4 Why This Problem is Hard ??**

---
##This isn't just about avoiding a storm; it’s a Data Integrity problem.


*   *No Ground Truth:* On Mars, there is no perfect weather station. The aircraft must create its own "truth" from conflicting sensor signals.

*   *Irreversible Consequences:* Unlike a software bug that can be rolled back, a wrong decision here results in "Lithobraking" (crashing into the planet).

# **2.1 System Architecture: The Triple-Layer Brain**
##To handle the 12-minute window, M.A.T.E.I. operates on three distinct logical layers:


---


*   *The Sentinel Layer (Data Fusion):* Collects raw data from LiDAR, Optical, and Barometric sensors. It calculates a Confidence Score.
*   *The Tactician Layer (Risk Assessment):* Compares the storm's severity against current battery levels and mission priority.
*   *The Executive Layer (Action):* Executes one of three protocols: Bypass, Emergency Grounding, or Stabilized Penetration.

# **2.2 The "Confidence Scoring" Logic**


---

On Mars, a single sensor can lie (a dusty lens). M.A.T.E.I. uses a Weighted Probability Model.


(w₁ × S_optical) + (w₂ × S_LiDAR) + (w₃ × S_baro)



Where w represents the reliability of the sensor in specific conditions. If the sensors disagree, M.A.T.E.I. triggers a "Data Validation" maneuver (changing altitude to see if the signal changes).




```python
import numpy as np

class MATEIPilot:
    def __init__(self, battery, mission_priority="High"):
        self.battery = battery  # Percentage
        self.priority = mission_priority
        self.status = "Normal"

    def evaluate_storm(self, optical_signal, lidar_signal, baro_signal):
        # Weighted Consensus: LiDAR is most trusted for distance, Baro for storm presence
        weights = {'opt': 0.2, 'lidar': 0.5, 'baro': 0.3}
        confidence = (optical_signal * weights['opt'] +
                      lidar_signal * weights['lidar'] +
                      baro_signal * weights['baro'])

        return round(confidence, 2)

    def decide_action(self, storm_intensity, confidence):
        print(f"Checking data... Confidence Level: {confidence*100}%")

        if confidence < 0.6:
            return "HOLD & SCAN: Data too noisy to decide."

        # Decision Matrix
        if storm_intensity > 0.8 and self.battery > 40:
            return "REROUTE: High intensity detected, battery sufficient for bypass."
        elif storm_intensity > 0.8 and self.battery <= 40:
            return "EMERGENCY LANDING: Risk of power failure during bypass."
        elif storm_intensity <= 0.3:
            return "PROCEED: Minor dust event, stabilizers active."
        else:
            return "EMERGENCY LANDING: Defaulting to safety protocol."

# Simulation
matei = MATEIPilot(battery=35, mission_priority="Critical")
# Let's simulate a high-intensity storm detected with high confidence
action = matei.decide_action(storm_intensity=0.9, confidence=0.85)
print(f"M.A.T.E.I. Decision: {action}")
```

# **2.4 Why This Approach?**


---



1.   *Resilience to False Positives:* By requiring a weighted consensus, we don't abort the mission just because a single sensor glitched.
1.   *Resource Awareness:* The system doesn't just look at the storm; it looks at its own "health" (battery). An autonomous system must know its own limits.
2.   *Deterministic Safety:* When confidence is low, the system defaults to the safest state (Landing), mirroring a "Fail-Safe" mechanism in enterprise data delivery.

# **3.1 The "Live Decision" Engine**

---



##Instead of a basic live-action animation, we are using Matplotlib combined with ipywidgets to build a robust, highly stable Data Analytics Dashboard. This approach allows us to instantly simulate the entire 12-minute decision window and generate a comprehensive mission report, demonstrating exactly how the algorithm processes noisy, conflicting data.

The Logic behind the Visualization:


*   *Tactical Navigation (The Environment)*: Visualizes the calculated flight path, the exact 14-minute Communication Blackout zone, and the "Cone of Uncertainty" representing the storm's shifting probability.

*   *Sensor Fusion (The Brain):* Highlights the core data challenge—extracting truth from noise. It plots the degraded signals (Optical and LiDAR) alongside the stable ones (Barometric) to compute the high-confidence M.A.T.E.I. Consensus line.
*   *Risk Radar (The Variables):* A multivariate polar chart that dynamically maps the storm's intensity against dust granularity and overall system uncertainty.


* *Data Pipeline Logs (The Outcome):* A structured telemetry output that explicitly justifies the drone's final autonomous strategy (e.g., Evasive Climb vs. Emergency Landing) based on the calculated confidence score.




# **3.2 The Simulation Code**

---



```python

# M.A.T.E.I. COCKPIT


import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects
import ipywidgets as widgets
from IPython.display import display, HTML
from dataclasses import dataclass
from datetime import datetime

# NOTEBOOK / COLAB SETTINGS

plt.rcParams["figure.dpi"] = 110
plt.rcParams["axes.titleweight"] = "bold"

# STYLE CONSTANTS

BG_MAIN = "#050505"
BG_PANEL = "#0a0a0a"
COLOR_CYAN = "#00FFC2"
COLOR_RED = "#FF0055"
COLOR_PURPLE = "#7000FF"
COLOR_ORANGE = "#FF9E00"
COLOR_OPTICAL = "#00e5ff"
COLOR_LIDAR = "#b2ff59"
COLOR_BARO = "#ff5252"
COLOR_GROUND_FILL = "#1a0f0a"
COLOR_GROUND_LINE = "#4d2612"

# DATA MODEL

@dataclass
class MissionConfig:
    storm_int: float
    batt: int
    wind_spd: int
    turbulence: float
    dust_grain: float
    payload_fragility: float
    seed: int = 42

# SIMULATION ENGINE

def compute_mission_state(cfg: MissionConfig):
    rng = np.random.default_rng(cfg.seed)
    x = np.linspace(0, 100, 500)

    # --- 1. TERRAIN & FLIGHT PATH ---
    ground = 15 * np.sin(x / 5) + 5 * np.cos(x / 2) + rng.normal(0, 1, size=x.size)
    path_alt = np.full_like(x, 250.0)
    noise = rng.normal(0, cfg.turbulence * 10, size=x.size)

    # --- 2. RISK & DECISION MODEL ---
    risk_score = (cfg.storm_int * 0.4 + (cfg.wind_spd / 120) * 0.3 + cfg.dust_grain * 0.3)
    status = "STABLE"
    path_color = COLOR_CYAN

    if risk_score > 0.6 or cfg.batt < 30:
        if cfg.payload_fragility > 0.7:
            status = "ULTRA-CONSERVATIVE LANDING"
            path_alt[150:300] = np.linspace(250, 20, 150)
            path_alt[300:] = 20
            path_color = COLOR_RED
        else:
            status = "DYNAMIC REROUTE"
            path_alt[200:] = 250 + (risk_score * 200)
            path_color = COLOR_PURPLE

    # --- 3. SENSOR FUSION DATA GENERATION ---
    t_sensor = np.linspace(0, 100, 200)
    approach = np.linspace(0, 1, 200) * cfg.storm_int

    optical_sig = np.clip(1.0 - approach - rng.normal(0, cfg.dust_grain * 0.4, 200), 0, 1)
    lidar_sig = np.clip(1.0 - (approach * 0.5) - rng.normal(0, cfg.dust_grain * 0.2, 200), 0, 1)
    baro_sig = np.clip(0.2 + (approach * 0.8) + rng.normal(0, 0.05, 200), 0, 1)

    confidence_array = (optical_sig * 0.15) + (lidar_sig * 0.35) + (baro_sig * 0.5)
    final_consensus = confidence_array[-1] * 100

    # --- 4. RADAR METRICS ---
    radar_labels = ["STORM PROB", "WIND", "DUST", "NOISE", "UNCERTAINTY"]
    radar_values = np.clip(np.array([
        cfg.storm_int,
        cfg.wind_spd / 120,
        cfg.dust_grain,
        (cfg.turbulence + cfg.dust_grain) / 2,
        1.0 - (final_consensus / 100)
    ]), 0, 1.2)

    return {
        "x": x, "ground": ground, "path_alt": path_alt, "noise": noise,
        "risk_score": risk_score, "status": status, "path_color": path_color,
        "t_sensor": t_sensor, "optical": optical_sig, "lidar": lidar_sig,
        "baro": baro_sig, "confidence": confidence_array, "consensus": final_consensus,
        "radar_labels": radar_labels, "radar_values": radar_values,
    }

# PLOTTING

def make_dashboard(cfg: MissionConfig):
    state = compute_mission_state(cfg)

    fig = plt.figure(figsize=(16, 10), facecolor=BG_MAIN, constrained_layout=True)
    gs = fig.add_gridspec(2, 2)

    ax_nav = fig.add_subplot(gs[0, 0])
    ax_fusion = fig.add_subplot(gs[0, 1])
    ax_radar = fig.add_subplot(gs[1, 0], polar=True)
    ax_telemetry = fig.add_subplot(gs[1, 1])

    # --- 1. TACTICAL NAVIGATION ---
    x = state["x"]
    path_color = state["path_color"]

    ax_nav.set_facecolor(BG_MAIN)
    ax_nav.fill_between(x, -100, state["ground"], color=COLOR_GROUND_FILL, alpha=0.85)
    ax_nav.plot(x, state["ground"], color=COLOR_GROUND_LINE, linewidth=1.0)

    # Glowing path
    rendered_path = state["path_alt"] + state["noise"]
    for i in range(1, 6):
        ax_nav.plot(x, rendered_path, color=path_color, alpha=0.10 / i, linewidth=1.8 * i)
    ax_nav.plot(x, rendered_path, color="white", linewidth=0.9, alpha=0.65)

    # 14-Min Comms Blackout Zone
    ax_nav.axvspan(15, 35, color='#ffffff', alpha=0.03, hatch='////')
    ax_nav.text(25, 550, "14-MIN COMM BLACKOUT", color='white', alpha=0.4, ha='center', fontsize=9, weight='bold')

    # Cone of Uncertainty
    cone_x = np.linspace(40, 100, 100)
    cone_upper = 250 + (cone_x - 40) * 5
    cone_lower = 250 - (cone_x - 40) * 5
    ax_nav.fill_between(cone_x, cone_lower, cone_upper, color=COLOR_ORANGE, alpha=cfg.storm_int * 0.15, label="Probability Cone")
    ax_nav.axvline(40, color=COLOR_ORANGE, linestyle=':', alpha=0.5)

    ax_nav.set_ylim(-50, 600)
    ax_nav.set_xlim(0, 100)
    ax_nav.set_title(f"TACTICAL NAV: {state['status']}", color=path_color, loc="left", fontsize=11)
    for spine in ax_nav.spines.values(): spine.set_color("#555")
    ax_nav.tick_params(colors="white")

    # --- 2. SENSOR FUSION & SIGNAL NOISE ---
    ax_fusion.set_facecolor(BG_PANEL)
    t = state["t_sensor"]

    ax_fusion.scatter(t, state["optical"], color=COLOR_OPTICAL, s=5, alpha=0.4, label="Optical (Noise)")
    ax_fusion.plot(t, state["lidar"], color=COLOR_LIDAR, linewidth=1, alpha=0.6, label="LiDAR (Degraded)")
    ax_fusion.plot(t, state["baro"], color=COLOR_BARO, linewidth=2, label="Barometric (Stable)")
    ax_fusion.plot(t, state["confidence"], color="white", linewidth=3, path_effects=[path_effects.withStroke(linewidth=5, foreground="#ffffff55")], label="M.A.T.E.I. Consensus")

    ax_fusion.set_title("SENSOR FUSION & NOISE FILTERING", color="white", fontsize=11)
    ax_fusion.set_ylim(0, 1.1)
    ax_fusion.legend(loc="lower left", facecolor=BG_PANEL, edgecolor="#555", labelcolor="white", fontsize=8)
    for spine in ax_fusion.spines.values(): spine.set_color("#555")
    ax_fusion.tick_params(colors="white", labelbottom=False)

    # --- 3. RISK RADAR ---
    labels = state["radar_labels"]
    values = np.r_[state["radar_values"], state["radar_values"][0]]
    angles = np.r_[np.linspace(0, 2 * np.pi, len(labels), endpoint=False), 0]

    ax_radar.set_facecolor(BG_MAIN)
    ax_radar.plot(angles, values, color=path_color, linewidth=2)
    ax_radar.fill(angles, values, color=path_color, alpha=0.28)
    ax_radar.set_xticks(angles[:-1])
    ax_radar.set_xticklabels(labels, color="white", fontsize=9)
    ax_radar.set_yticklabels([])
    ax_radar.set_ylim(0, 1.2)
    ax_radar.grid(alpha=0.25)

    # --- 4. TELEMETRY & LOGS ---
    ax_telemetry.set_facecolor(BG_PANEL)
    ax_telemetry.axis("off")

    now_str = datetime.now().strftime("%M:%S:%f")[:-3]
    telemetry = (
        f"[M.A.T.E.I. DATA PIPELINE]\n"
        f"--------------------------\n"
        f"TIME_REF: T+{now_str}\n"
        f"BATTERY: {cfg.batt}%\n"
        f"DATA_NOISE_LEVEL: {cfg.dust_grain * 100:.1f}%\n"
        f"SENSOR_CONSENSUS: {state['consensus']:.1f}%\n"
        f"SYSTEM_UNCERTAINTY: {(100 - state['consensus']):.1f}%\n"
        f"RISK_SCORE: {state['risk_score']:.3f}\n"
        f"--------------------------\n"
        f"EARTH_LINK: SEVERED\n"
        f"AUTONOMOUS_STRATEGY: {state['status']}"
    )

    ax_telemetry.text(0.04, 0.5, telemetry, color=path_color, family="monospace", fontsize=11, va="center", weight="bold")

    return fig

# RENDER CALLBACK

def render_dashboard(storm_int, batt, wind_spd, turbulence, dust_grain, payload_fragility, seed):
    cfg = MissionConfig(
        storm_int=storm_int,
        batt=batt,
        wind_spd=wind_spd,
        turbulence=turbulence,
        dust_grain=dust_grain,
        payload_fragility=payload_fragility,
        seed=seed,
    )
    fig = make_dashboard(cfg)
    plt.show()
    plt.close(fig)

# WIDGETS

out = widgets.Output()
style = {"description_width": "150px"}
wide = widgets.Layout(width="95%")

w_storm = widgets.FloatSlider(value=0.8, min=0, max=1.0, step=0.01, description="Storm Prob.", style=style, layout=wide, continuous_update=False)
w_wind = widgets.IntSlider(value=65, min=0, max=120, step=1, description="Wind Speed (m/s)", style=style, layout=wide, continuous_update=False)
w_turb = widgets.FloatSlider(value=0.4, min=0, max=1.0, step=0.01, description="Atmo. Turbulence", style=style, layout=wide, continuous_update=False)
w_dust = widgets.FloatSlider(value=0.7, min=0, max=1.0, step=0.01, description="Signal Noise (Dust)", style=style, layout=wide, continuous_update=False)
w_batt = widgets.IntSlider(value=80, min=0, max=100, step=1, description="Battery %", style=style, layout=wide, continuous_update=False)
w_frag = widgets.FloatSlider(value=0.3, min=0, max=1.0, step=0.01, description="Payload Value", style=style, layout=wide, continuous_update=False)
w_seed = widgets.IntSlider(value=42, min=0, max=9999, step=1, description="Simulation Seed", style=style, layout=wide, continuous_update=False)

run_btn = widgets.Button(description=" RUN DATA PIPELINE & PREDICT", layout=widgets.Layout(width="100%", height="52px"))
run_btn.style.button_color = "#111111"

controls_env = widgets.VBox([widgets.HTML("<b style='color:white'>ENVIRONMENT & NOISE</b>"), w_storm, w_wind, w_turb, w_dust])
controls_hw = widgets.VBox([widgets.HTML("<b style='color:white'>AIRCRAFT & MISSION</b>"), w_batt, w_frag, w_seed])

tabs = widgets.Tab(children=[controls_env, controls_hw])
tabs.set_title(0, "Data Signals")
tabs.set_title(1, "Hardware")

def refresh(_=None):
    with out:
        out.clear_output(wait=True)
        render_dashboard(w_storm.value, w_batt.value, w_wind.value, w_turb.value, w_dust.value, w_frag.value, w_seed.value)

run_btn.on_click(refresh)

display(HTML("<h1 style='color:#00FFC2; font-family:monospace; text-align:center;'>M.A.T.E.I. DATA FUSION ENGINE</h1>"))
display(tabs, run_btn, out)
refresh()
```
